```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,root_mean_squared_error,mean_squared_error,r2_score
```

```python
df=pd.read_csv('Student_Performance1.csv')
df.head()
```

```python
df['Extracurricular Activities']=df['Extracurricular Activities'].map({'No':0,'Yes':1})
```

```python
x=df.drop('Performance Index',axis=1)# x used be in 2-D always
y=df['Performance Index']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8)
```

```python
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred_train=model.predict(xtest)
ypred_train
```

```python
ytrain-ypred_train
```

```python
mse=mean_squared_error(ytrain,ypred_train)
mse
```

```python
mae=mean_absolute_error(ytrain,ypred_train)
mae
```

```python
rmse=root_mean_squared_error(ytrain,ypred_train)
rmse
```

```python
print(f"mae:{mae}\nrmse:{rmse}")
```

```python
r2_scorce=r2_score(ytrain,ypred_train)
r2_scorce
```
