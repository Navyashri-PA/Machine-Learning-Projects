```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,root_mean_squared_error,mean_squared_error,r2_score
```

# Analysizing Data

```python
df=pd.read_csv("Ice_cream selling data.csv")
df.head()
```

```python
x=df[['Temperature (°C)']]
y=df['Ice Cream Sales (units)'] 
```

# Checking if data is linear or not

```python
plt.scatter(x,y)
```

```python
df.corr()
```

# Data is non-linear, Hence we cannot apply linear regression

```python
xtrain, xtest, ytrain, ytest =train_test_split(x,y,train_size=0.8)
```

```python
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred_train=model.predict(xtrain)
ypred_train
```

```python
ytrain-ypred_train
```

```python
mse=mean_squared_error(ytrain,ypred_train)
mse
```

```python
mae=mean_absolute_error(ytrain,ypred_train)
mae
```

```python
rmse=root_mean_squared_error(ytrain,ypred_train)
rmse
```

```python
print(f"mae:{mae}\nrmse:{rmse}")
```

```python
r2_scorce=r2_score(ytrain,ypred_train)
r2_scorce
```
