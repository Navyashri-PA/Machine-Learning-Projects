```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.model_selection import train_test_split
# from sklearn.metrics import mean_absolute_error,root_mean_squared_error,mean_squared_error,r2_score
```

```python
df=pd.read_csv("Ice_cream selling data.csv")
df.head()
```

```python
x=df[['Temperature (°C)']]
y=df['Ice Cream Sales (units)'] 
```

```python
plt.scatter(x,y)
```

```python
xtrain, xtest, ytrain, ytest =train_test_split(x,y,train_size=0.8, random_state=42)
```

```python
plt.scatter(xtrain,ytrain)
```

```python
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
ypred=model.predict(xtrain)
ypred
```

```python
plt.scatter(xtrain,ytrain)
plt.scatter(xtrain,ypred)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=1)
xtrain_poly= poly.fit_transform(xtrain)
xtrain_poly
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
xtrain_poly= poly.fit_transform(xtrain)
xtrain_poly
```

```python
model= LinearRegression()
model.fit(xtrain_poly,ytrain)
model.score(xtrain_poly,ytrain)
```

```python
ypred= model.predict(xtrain_poly)
```

```python
xtrain_poly[:,1]
```

```python
plt.scatter(xtrain_poly[:,1],ytrain) #actual input, actual output
plt.scatter(xtrain_poly[:,1],ypred) #actual input, predicted output
```
