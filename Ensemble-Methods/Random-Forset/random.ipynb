```python
from sklearn.datasets import load_diabetes
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import r2_score,mean_absolute_error,root_mean_squared_error,mean_squared_error
```

```python
df=load_diabetes(as_frame=True)
```

```python
x=df.data
y=df.target
```

```python
df.data
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
cv=GridSearchCV(estimator=RandomForestRegressor(),param_grid={
    'n_estimators':[100,500,300,400],
    'criterion':['squared_error','absolute_error'],
    'min_samples_split':[2,3,5,7],
    'min_samples_leaf':[1,2,3,7],
},cv=3,scoring="r2",error_score='raise')
```

```python
cv.fit(xtrain,ytrain)
```

```python
para=cv.best_params_
para
```

```python
model=RandomForestRegressor(n_estimators=300,random_state=42)
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred_train=model.predict(xtrain)
ypred_test=model.predict(xtest)
```

```python
r2_train=r2_score(ytrain,ypred_train)
r2_train
```

```python
r2_test=r2_score(ytest,ypred_test)
r2_test
```

```python
r2_train=r2_score(ytrain,cv_train)
r2_train
```

```python
r2_test=r2_score(ytest,cv_test)
r2_test
```
