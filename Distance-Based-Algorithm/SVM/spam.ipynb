```python
import pandas as pd
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
df=pd.read_csv('spam.csv',encoding_errors='ignore')
df.head()
```

```python
df.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'],inplace=True)
```

```python
df.head()
```

```python
x=df['v2']
y=df['v1']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
pipeline=Pipeline(steps=[
    ('count',CountVectorizer()),
    ('model',MultinomialNB())
])
pipeline.fit(xtrain,ytrain)
```

```python
pipeline1=Pipeline(steps=[
    ('count',TfidfVectorizer()),
    ('model',MultinomialNB())
])
pipeline1.fit(xtrain,ytrain)
```

```python
pipeline2=Pipeline(steps=[
    ('count',TfidfVectorizer()),
    ('model',LogisticRegression())
])
pipeline2.fit(xtrain,ytrain)
```

```python
ypred_xtrain=pipeline.predict(xtrain)
ypred_xtest=pipeline.predict(xtest)
```

```python
ypred_xtrain1=pipeline1.predict(xtrain)
ypred_xtest1=pipeline1.predict(xtest)
```

```python
ypred_xtrain2=pipeline2.predict(xtrain)
ypred_xtest2=pipeline2.predict(xtest)
```

# multinomial and countVector

```python
print(classification_report(ytrain,ypred_xtrain))
print(classification_report(ytest,ypred_xtest))
```

# multinomial and Tfidf

```python
print(classification_report(ytrain,ypred_xtrain1))
print(classification_report(ytest,ypred_xtest1))
```

```python
print(classification_report(ytrain,ypred_xtrain2))
print(classification_report(ytest,ypred_xtest2))
```
